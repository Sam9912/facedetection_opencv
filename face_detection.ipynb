```python
jd=cv2.imread('jd1.jpg',0)
wr=cv2.imread('wr1.jpg',0)
jw=cv2.imread('jw.jpg',0)
```

```python
plt.imshow(jw,cmap='gray')
```

```python
classifier=cv2.CascadeClassifier('C:/Users/DELL/Desktop/sam/opencv project/haarcascade files/haarcascade_frontalface_default.xml')
```

```python
def face_detect(img):
    image=img.copy()
    face_rect=classifier.detectMultiScale(image)
    for (x,y,w,h) in face_rect:
        cv2.rectangle(image,(x,y),(x+w,y+h),(225,10,212),12)
    return image
```

```python
check_image=face_detect(jw)
```

```python
plt.imshow(check_image,cmap='gray')
```

```python
def face_detect2(img):
    image=img.copy()
    face_rect=classifier.detectMultiScale(image,scaleFactor=1.1, minNeighbors=2)
    for (x,y,w,h) in face_rect:
        cv2.rectangle(image,(x,y),(x+w,y+h),(225,10,212),12)
    return image
```

```python
check_image=face_detect(jw)
```

```python
plt.imshow(check_image,cmap='gray')
```

# realtime face detection


```python
cam=cv2.VideoCapture(0)
while True:
    ret, frame=cam.read()
    frame=face_detect2(frame)
    cv2.imshow('camera',frame)
    
    close=cv2.waitKey(1)
    if close == 27:
        break        
cam.release()
cv2.destroyAllWindows()
```
